# Grafico de Notas

```python
from random import randrange, seed
```

```python
seed(11)
```

```python
randrange(0,11)
```

```python
notas_bimestre = []

for notas in range(8):
    notas_bimestre.append(randrange(0,11))

notas_bimestre
```

```python
len(notas_bimestre)
```

# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib import markers
x = list(range(1,9))
y = notas_bimestre
plt.plot(x,y, marker= 'o')
plt.title('Grafico de Notas')
plt.xlabel('Notas')
plt.xlabel('Provas')
plt.show()
```
